```python
import cv2 as cv
import os
from matplotlib import pyplot as plt
import numpy as np
from PIL import Image
import colorsys
def create_hls_array(image):
    pixels = image.load()
    hls_array = numpy.empty(shape=(image.height, image.width, 3), dtype=float)
    for row in range(0, image.height):
        for column in range(0, image.width):
            rgb = pixels[column, row]
            hls = colorsys.rgb_to_hls(rgb[0]/255, rgb[1]/255, rgb[2]/255)
            hls_array[row, column, 0] = hls[0]
            hls_array[row, column, 1] = hls[1]
            hls_array[row, column, 2] = hls[2]
    return hls_array
SAMPLE1_PATH = '../data/empire_sample_2.png'
SAMPLE2_PATH = '../data/temp_result/2580.png'
SAMPLE3_PATH = '../data/temp_result/2592.png'
SAMPLE4_PATH = '../data/temp_result/1113.png'
SAMPLE5_PATH = '../data/temp_result/2000.png'
SAMPLE6_PATH = '../data/temp_result/2600.png'
SAMPLE7_PATH = '../data/temp_result/1234.png'
SAMPLE8_PATH = '../data/temp_result/354.png'
```

```python
img = Image.open(SAMPLE1_PATH)
img1 = img.transpose(Image.ROTATE_90)
img2 = create_hls_array(Image.open(SAMPLE1_PATH))
img = img2#(img+img2)/2#np.array(img)
img1 = np.array(img1)
plt.figure(figsize=(18,18))
img1[:,:,2] *= 0
img1[:,:,0] *= 0
plt.imshow(img)
plt.show()
plt.imshow(img[:,:, 0])
plt.show()
plt.imshow(img[:,:, 1])
plt.show()
plt.imshow(img[:,:, 2])
plt.show()
print(img.shape)
img_sum = img[:,:,0]+img[:,:,1]+img[:,:,2]
way = 1
sum_0 = np.sum(img_sum, way)
sum_1 = np.sum(img[:,:, 0], way)
sum_2 = np.sum(img[:,:, 1], way)
sum_3 = np.sum(img[:,:, 2], way)
x = range(0,100)
#hist, bin_edges = np.histogram(sum_0, density=True)
plt.plot(x, sum_0)
plt.show()
plt.plot(x, sum_1)
plt.show()
plt.plot(x, sum_2)
plt.show()
plt.plot(x, sum_3)
plt.show()
```

# Erosion & Dilation

```python
kernel = np.ones((4,4),np.uint8)

dilation = cv.dilate(img_aft3,kernel,iterations = 1)
erosion = cv.erode(dilation,kernel,iterations = 1)

dilation = cv.dilate(erosion,kernel,iterations = 1)
erosion = cv.erode(dilation,kernel,iterations = 1)

dilation = cv.dilate(erosion,kernel,iterations = 1)
erosion = cv.erode(dilation,kernel,iterations = 1)

dilation = cv.dilate(erosion,kernel,iterations = 1)
erosion = cv.erode(dilation,kernel,iterations = 1)
erosion2 = cv.erode(dilation,kernel,iterations = 2)

#erosion = cv.erode(erosion,kernel,iterations = 1)
#erosion = cv.erode(erosion,kernel,iterations = 1)
#erosion = cv.erode(dilation,kernel,iterations = 1)
ret, erosion2_thresh = cv.threshold(erosion,0,255,cv.THRESH_BINARY_INV+cv.THRESH_OTSU)
#ret, thresh = cv.threshold(gray,0,255,cv.THRESH_OTSU)
plt.imshow(erosion2)
plt.show()
plt.imshow(erosion2_thresh)
plt.show()
```
